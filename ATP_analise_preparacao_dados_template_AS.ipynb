### 4.2 Variância dos dados.

Criamos uma função para calcular a variância dos dados númericos.

Como descrito, na hipótse anteriormente, confirmamos que exintem dados dispersos em relação a média.
Possivelmente existem outliers que desviam significativamente do padrão geral.

````
def calcular_variancia_numerica(aux):

    dados_numericos = aux.select_dtypes(include='number')
    variancia = dados_numericos.var()
    return variancia

variancia_numerica = calcular_variancia_numerica(df)
print(variancia_numerica)
````

### 5) Utilizamos a função abaixo para calcular a quantidade de dados em cada variavel categórica.

````
def calcular_variavel_categorica(aux):

    dados_categoricos = aux.select_dtypes(include=['object','bool'])
    total = dados_categoricos.count()
    return total

variancia_numerica = calcular_variavel_categorica(df)
print(variancia_numerica)
````

##Análise Univariada:  **Salário (Salary)**

````
plt.figure(figsize=(15,5))
sns.set_theme(style="whitegrid", palette="dark")
sns.histplot(x='salary', data=df, kde=True)
sns.histplot(x='salary', data=df, kde=True).axvline(x=df['salary'].mean(), color='w',linestyle="--")
plt.ticklabel_format(style='sci', axis='x', scilimits=(3,6))
plt.tight_layout()
````

Podemos observar que a distribuição esta concentrada nos valores próximos a média, no entanto devemos verificar os **outliers** que possivelmente pertencem a cargos da alta direção/presidência e estão inviabilizando a análise completa dos dados.

Abaixo calculamos os valores máximo e média dos salários para entendermos a discrepância encontrada.

````
format(df['salary'].max(),'.2f')
````

````
format(df['salary'].mean(),'.2f')
````

A título de exercício iremos limitar a quantidade de salários a fim de verificar se teremos uma melhor distribuição dos dados.

````
plt.figure(figsize=(15,5))
sns.rugplot(x='salary', data=df[df['salary']<1e6], height=.03, color='darkred').set_title("Distribuição de salários")
sns.histplot(x='salary', data=df[df['salary']<1e6], kde=True).axvline(x=df['salary'].mean(), color='r',linestyle="--")
plt.ticklabel_format(style='sci', axis='x', scilimits=(2,1))
````

Após limitado os valores (exclusão dos outliers) verificamos uma distribuição  simétrica dos dados.

Podemos concluir que grande parte dos funcionários estão próximos de uma faixa salarial de 2,5x10^5.

A média dos valores esta sendo fortemente influenciada devido aos valores discrepantes dos outliers.

##Análise Univariada:  **Despesas (expenses)**

````
plt.figure(figsize=(8,5))
sns.set_theme(style="ticks", palette="Set1")
sns.boxplot(y="expenses",data=df)
sns.despine(offset=10, trim=True)
````

Verificamos no gráfico acima que os outliers impossibilitam a análise dos dados.

Dessa forma vamos limitar os valores.

````
df1=df[df['expenses']<1.7e5]
plt.figure(figsize=(8,5))
sns.set_theme(style="ticks", palette="Set3")
sns.boxplot(y="expenses",data=df1)
#sns.despine(offset=21, trim=True)
plt.figure(figsize=(15,5))
sns.displot(data=df1, x="expenses", kde=True)
sns.histplot(x='expenses', data=df[df['expenses']<1.7e5], kde=True).axvline(x=df['expenses'].mean(), color='r',linestyle="--")
plt.ticklabel_format(style='sci', axis='x', scilimits=(2,1))



````

Após retirada dos outliers, conseguimos visualizar melhor o boxplot juntamente com o histograma e verificar a distribuição assimétrica em que os dados aparecem com maior frequência a medida que os valores são menores.
